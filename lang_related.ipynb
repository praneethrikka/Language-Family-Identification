```python
# stan model
stan_program = """
functions{

  matrix cov_GPL2(vector x, real sq_eta, real sq_rho, real delta) {
    int N = dims(x)[1];
    matrix[N, N] K;
    for (i in 1:(N-1)) {
      K[i, i] = sq_eta + delta;
      for (j in (i + 1):N) {
        K[i, j] = sq_eta * exp(-(square(x[i]-x[j])/sq_rho));
        K[j, i] = K[i, j];
      }
    }
    K[N, N] = sq_eta + delta;
    return K;
  }
}
data {
  int N;
  vector[N] T;
  vector[N] S;
  matrix[N,N] M;
}
parameters {
  real<lower=0> rhosq;
  real<lower=0> etasq;
  real<lower=0> rhosq_geo;
  real<lower=0> etasq_geo;
  real<lower=0> sigma;
  real a;
  vector[N] gp;

}
transformed parameters {
  matrix[N,N] SIGMA;

  SIGMA = cov_GPL2(T, etasq, rhosq, 0.01);


}
model{
  vector[N] mu;
   mu = a + gp ;

  S ~ normal(mu, sigma);

  gp ~ multi_normal(rep_vector(0,N),SIGMA);
  a ~ normal(0,1);
  rhosq ~ exponential( 0.5 );
  etasq ~ exponential( 0.5 );
  sigma ~ exponential( 1 );

}
"""

datlist = {
    'N' :100,
    'T' : list(df['new_feature'][:100]),
    'S' : list(range(100)),
    'M' : dist
}


mod =  stan.build(stan_program, data = datlist) # building model
fit = mod.sample(num_chains = 4, num_samples = 500) # fitting model
```
